```python
def get_px(train, val, col):
    df = train
    last = df.columns[len(df.columns) - 1]
    yes = df['BUY'].value_counts()['YES']
    no = df['BUY'].value_counts()['NO']
    df = train
    p_x_and_yes = len(df[(df[col] == val) & (df['BUY'] == 'YES')])
    p_x_and_no = len(df[(df[col] == val) & (df['BUY'] == 'NO')])
    p_x_given_yes = p_x_and_yes/yes
    p_x_given_no = p_x_and_no/no
    
    return p_x_and_yes, p_x_and_no, p_x_given_yes, p_x_given_no
```

```python
def mul(arr):
    ans = 1
    for val in arr:
        ans *= val
    return ans


def print_arr_mul(arr):
    print('{:.4f}'.format(arr[0]), end = '')
    for i in range(1, len(arr)):
        print(' * ' + '{:.4f}'.format(arr[i]), end = ' ')
        
    #print(' = ' + '{:.4f}'.format(mul(arr)))
    print(' = ' + str(mul(arr)))
a = [1.2324, 3.6633333, 363.3239953]
print_arr_mul(a)
```

```python
p_arr = []
```

```python
def compute_ass3(train, test, p_arr):
    n = len(train)

    yes = 29
    no = 21

    pyes = yes / n
    pno = no / n
    
    df = train
    
    for col in df.columns:
        print(col)
        for val in df[col].unique():
            #P(Xi | YES) 
            #p_x_and_yes = len(df[(df[col] == val) & (df['BUY'] == 'YES')])
            #p_x_and_no = len(df[(df[col] == val) & (df['BUY'] == 'NO')])
            #p_x_given_yes = p_x_and_yes/yes
            #p_x_given_no = p_x_and_no/no
            p_x_and_yes, p_x_and_no, p_x_given_yes, p_x_given_no = get_px(df, val, col)
            print('P(' + str(val) +' | YES) = ' + str(p_x_and_yes) + '/' + str(yes) + '\t= ' + "{:.4f}".format(p_x_given_yes) )
            print('P(' + str(val) +' | NO)  = ' + str(p_x_and_no) + '/' + str(no) + '\t= ' + "{:.4f}".format(p_x_given_no))
            print()
        print()
    
    print('---------------------test data result------------------------------------\n\n\n')
    
    df_test = test
    
    for row in range(len(df_test)):
        arr_yes = []
        arr_no = []
        for features in test.columns:
            val = test[features][row]
            p_x_and_yes, p_x_and_no, p_x_given_yes, p_x_given_no = get_px(df, val, features)
            arr_yes.append(p_x_given_yes)
            arr_no.append(p_x_given_no)
        
        p_x_tot_given_yes = mul(arr_yes)
        p_x_tot_given_no = mul(arr_no)
        
        print('P(X' + str(row) + ' | YES) = ', end = '')
        print_arr_mul(arr_yes)
        final_yes = mul(arr_yes) * pyes
        print('P(X' + str(row) + ' | YES) * P(YES) = ' + str(final_yes), end = '\n')
        
            
        print('P(X' + str(row) + ' | NO) = ', end = '')
        print_arr_mul(arr_no)
        final_no = mul(arr_no) * pno
        print('P(X' + str(row) + ' | NO) * P(NO) = ' + str(final_no), end = '\n')
        
        print()
        #predicition = final_yes > final_no ? 'YES' : 'NO';
        prediction = 'YES' if final_yes > final_no else 'NO'
        print('Predicition: ' + prediction)
        
        #np.append(p_arr, prediction)
        p_arr.append(prediction)
        print()
        
        
        
            

        
compute_ass3(df, test_unlabeled, p_arr)
```
